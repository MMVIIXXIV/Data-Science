**Form worldbank_gdp.csv answer the following questions.

1.Which country GDP is growing during the year 2010 and year 2018 .Support answer with visualizations.**

```python
world_gdp   =pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Second year first sem/DATA SCIENCE/W3/W3 - Data Sets & Solutions/WorldBank_GDP.csv")
china_gdp   =world_gdp[world_gdp["Country Name"]=="China"][["Year","GDP"]]
germany_gdp =world_gdp[world_gdp["Country Name"]=="Germany"]
usa_gdp     =world_gdp[world_gdp["Country Name"]=="United States"]
jp_dgp      =world_gdp[world_gdp["Country Name"]=="Japan"]
plt.plot    (   china_gdp["Year"]      ,china_gdp["GDP"]    ,label="China")
plt.plot    (   germany_gdp["Year"]    ,germany_gdp["GDP"]  ,label="Germany")
plt.plot    (   usa_gdp["Year"]        ,usa_gdp["GDP"]      ,label="United States")
plt.plot    (   jp_dgp["Year"]         ,jp_dgp["GDP"]       ,label="Japan")
plt.xlabel  ("Year")
plt.ylabel  ("GDP")
plt.legend  ()
plt.show    ()

```

The data from graph shows GDP of United States and China are growing during the year 2010 and 2018.
