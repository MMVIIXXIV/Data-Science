```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Read the CSV file into a DataFrame
music_df = pd.read_csv('music_unclean.csv', index_col=0)

# Drop rows with missing values in specific columns
music_df = music_df.dropna(subset=['genre', 'popularity', 'loudness', 'liveness', 'tempo'])

# Encode the 'genre' column as binary labels
music_df['genre'] = np.where(music_df['genre'] == 'Rock', 1, 0)

# Separate features (X) and the target variable (y)
X = music_df.drop('genre', axis=1).values
y = music_df['genre'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Impute missing values using SimpleImputer
imputer = SimpleImputer()
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Train a Logistic Regression model
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train_imputed, y_train)

# Make predictions on the test data
y_pred = logistic_regression.predict(X_test_imputed)

# Calculate and print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

# Load the music_df from music_unclean.csv
music_df = pd.read_csv('music_unclean.csv')

# Print the number of missing values for each column in ascending order
missing_values = music_df.isnull().sum().sort_values(ascending=True)
print("Missing Values:")
print(missing_values)

# Remove columns with 50 or fewer missing values
threshold = 50
music_df = music_df.dropna(thresh=len(music_df) - threshold, axis=1)

# Convert music_df["genre"] to binary labels (1 for "Rock" and 0 for others)
music_df['genre'] = np.where(music_df['genre'] == 'Rock', 1, 0)

# Define features (X) and the target variable (y)
X = music_df.drop('genre', axis=1)
y = music_df['genre']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instantiate an imputer to handle missing values
imputer = SimpleImputer()

# Fit and transform the imputer on the training data
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Create a DataFrame to check for missing values after imputation
columns = X.columns  # Use the columns from X, which have been reduced
check = pd.DataFrame(X_train_imputed, columns=columns)
print("Missing Values After Imputation:")
print(check.isna().sum().sort_values())

# Instantiate a KNN classifier with three neighbors
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the KNN classifier to the training data
knn.fit(X_train_imputed, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test_imputed)

# Calculate and print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
